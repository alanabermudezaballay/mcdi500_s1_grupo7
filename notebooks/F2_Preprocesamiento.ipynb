```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración global de estilos para gráficos
sns.set_theme(style="whitegrid")

print("Librerías importadas y entorno configurado correctamente.")
```

# III. Obtención de Datos

```python
# Definición de la función para mayor modularidad
def cargar_datos(nombre_archivo):
    ruta = Path.cwd().parent / "data" / "raw" / nombre_archivo
    print("Ruta usada:", ruta)
    return pd.read_csv(ruta)

# Ejecución de la carga
df = cargar_datos("predictive_maintenance_dataset.csv")

print("Dataset cargado satisfactoriamente.")
display(df.head())
```

# IV. Diagnóstico de Calidad

```python
# Resumen estadístico
print("Resumen estadístico de los datos:")
display(df.describe())
```

```python
# Diagnóstico adicional de calidad
print("--- Valores Nulos ---")
print(df.isnull().sum())

print("\n--- Duplicados ---")
print(f"Total registros duplicados: {df.duplicated().sum()}")

print("\n--- Tipos de Datos ---")
print(df.dtypes)
```

```python
from pathlib import Path

print("Ruta actual de trabajo:")
print(Path.cwd())
```

```python
print("Cantidad de filas:", df.shape[0])

print("\nCantidad de duplicados:")
print(df.duplicated().sum())

print("\nRegistros duplicados:")
display(df[df.duplicated()])
```

```python
# Crear copia antes (buena práctica)
df_before = df.copy()

# Eliminar duplicados
df = df.drop_duplicates()

print("Duplicados eliminados correctamente.")
```

```python
print("Filas antes:", df_before.shape[0])
print("Filas después:", df.shape[0])

print("\nDuplicados restantes:")
print(df.duplicated().sum())
```

```python
# Ver tipo antes
print("Tipo antes:")
print(df["date"].dtype)

# Convertir a datetime
df["date"] = pd.to_datetime(df["date"])

# Ver tipo después
print("\nTipo después:")
print(df["date"].dtype)
```

```python
# Crear nuevas variables de tiempo
df["year"] = df["date"].dt.year
df["month"] = df["date"].dt.month
df["day"] = df["date"].dt.day

print("Variables temporales creadas correctamente.")
```

```python
display(df.head())
```

```python
print("VALIDACIÓN FINAL DEL DATASET\n")

print("Filas iniciales:", df_before.shape[0])
print("Filas finales:", df.shape[0])

print("\nDuplicados restantes:")
print(df.duplicated().sum())

print("\nTipos de datos finales:")
print(df.dtypes)

print("\nResumen estadístico:")
display(df.describe())
```

```python
# Crear carpeta processed si no existe
ruta_carpeta = Path.cwd().parent / "data" / "processed"
ruta_carpeta.mkdir(parents=True, exist_ok=True)

# Ruta del archivo final
ruta_salida = ruta_carpeta / "dataset_procesado.csv"

# Guardar archivo
df.to_csv(ruta_salida, index=False)

print("Dataset exportado correctamente en:")
print(ruta_salida)
```
